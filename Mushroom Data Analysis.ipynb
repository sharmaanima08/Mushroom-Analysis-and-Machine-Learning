# Data Cleaning and Preprocessing

```python
'''1. Handling missing values
   – Mean / Median / Mode / Drop / Fill
2. Standardizing column names
3. Duplicate analysis (why kept / removed)
4. Outlier handling (IQR / Z-score / capping / removal)'''
```

```python
for col in ['ring-type','cap-surface','gill-attachment']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
for col in ['gill-spacing','stem-surface','stem-root']:
    data[col].fillna("Unknown", inplace=True)
```

```python
data.drop(columns=["veil-type", "veil-color", "spore-print-color"], inplace=True)
```

```python
data.columns=(
    data.columns
    .str.strip()
    .str.lower()
    .str.replace("-" , "_"))
data.rename(columns={'class' : 'mushroom_class'}, inplace=True)
```

```python
duplicates= data[data.duplicated(keep=False)]
duplicates.head()
```

```python
for col in data.columns:
    duplicate_count = data[col][data.duplicated(subset=[col], keep=False)].count()
    print(f"{col}: {duplicate_count} duplicates")
```

```python
num_col = ['cap_diameter', 'stem_height', 'stem_width']
for col in num_col:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    lower = max(q1 - 1.5 * iqr , 0)
    upper = q3 + 1.5 * iqr

    data[col] = np.where(data[col] < lower , lower , np.where(data[col] > upper , upper , data[col]))
    print(f"{col} : values capped between {lower} and {upper}")
```

```python
'''Insights:
. Columns ring_type, cap_surface, and gill_attachment were filled using mode due to low to moderate missing values.
. Columns gill_spacing, stem_surface, and stem_root were filled with 'Unknown' because of high missing-value proportions.
. Columns veil_type, veil_color, and spore_print_color were dropped due to extremely high missing values.
. Column names were standardized by replacing hyphens (-) with underscores (_), and the target column 'class' was renamed to 'mushroom_class'.
. Duplicate rows were not removed, as repetition is expected due to the categorical nature of the dataset.
. Outliers in numerical columns were handled using the capping method to reduce skewness without data loss.'''
```

# Statistical and Relationship EDA

```python
'''1. Univariate analysis
2. Bivariate analysis
3. Multivariate analysis
4. Statistical measures
   – Variance
   – Standard deviation
   – Skewness
   – Kurtosis'''
```

```python
def univariate_analysis(col, data):
    if data[col].dtype.kind in 'if': 
        summary = {
            'column': col,
            'type': 'numeric',
            'mean': data[col].mean(),
            'median': data[col].median(),
            'mode': data[col].mode(),
            'min': data[col].min(),
            'max': data[col].max(),
            'range': data[col].max() - data[col].min(),
            'skewness': data[col].skew(),
            'unique_vals': data[col].nunique()
        }
    else: 
        vc = data[col].value_counts()
        summary = {
            'column': col,
            'type': 'categorical',
            'mode': vc.idxmax(),
            'top_count': vc.max(),
            'unique_vals': data[col].nunique(),
            'top_pct': (vc.max() / len(data)) * 100
        }

    return pd.DataFrame([summary])
```

```python
df = pd.concat([univariate_analysis(col, data) for col in data.columns],
                        ignore_index=True)
df
```

```python
str_pos = 0.7
str_neg = -0.7
corr=data.select_dtypes(include=['int64','float64']).corr()
strong_positive=(corr.where((corr > str_pos) & (corr < 1.0)).stack().reset_index())
strong_negative=(corr.where(corr < str_neg).stack().reset_index())

strong_positive.columns = ['Feature 1','Feature 2','Correlation']
strong_negative.columns = ['Feature 1','Feature 2','Correlation']

print(f"Strong Positive Correlation {strong_positive}")
print(f"Strong Negative Correlation {strong_negative}")
```

```python
'''Insights:
. A correlation heatmap was generated for numerical features.
. No strong positive or negative correlations were observed, indicating absence of multicollinearity among numerical variables.'''
```
